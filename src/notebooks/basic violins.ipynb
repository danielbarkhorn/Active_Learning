```python
active_svm = pickle.load(open('../tests/activeSVMF1s.p','rb'))
rand_svm = pickle.load(open('../tests/randSVMF1s.p','rb'))
active_rf = pickle.load(open('../tests/activeRFF1s.p','rb'))
rand_rf = pickle.load(open('../tests/randRFF1s.p','rb'))
```

```python
svms = [rand_svm, active_svm]
rfs = [rand_rf, active_rf]
```

```python
axes = figure(figsize=(10,10)).add_subplot(111)
violinplot(svms,showmedians=True)
plt.setp(axes, xticks=[y+1 for y in range(len(svms))],
         xticklabels=['Random', 'Active'])
tick_params(axis='x', which='major', labelsize=15)
tick_params(axis='y', which='major', labelsize=15)
ylabel('F1 score',size=20)
title('Support Vector Classifier',fontsize=20);
```

```python
axes = figure(figsize=(10,10)).add_subplot(111)
violinplot(rfs,showmedians=True)
plt.setp(axes, xticks=[y+1 for y in range(len(rfs))],
         xticklabels=['Random', 'Active'])
tick_params(axis='x', which='major', labelsize=15)
tick_params(axis='y', which='major', labelsize=15)
ylabel('F1 score',size=20)
title('Random Forest Classifier',fontsize=20);
```
